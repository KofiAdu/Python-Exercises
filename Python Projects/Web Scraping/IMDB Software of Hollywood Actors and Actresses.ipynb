```python
import numpy as np
import pandas as pd
import requests
import csv
from urllib.request import urlopen
from os import path
import os
import sys
import json
import re
from bs4 import BeautifulSoup

class ActorsnActresses(object):

    def __init__(self):
        super(ActorsnActresses, self).__init__()
        self.file_name=None


    def printingcsv(self, file_name):
        df = pd.read_csv(file_name, encoding = "ISO-8859-1")
        #print(df)
        return df
    def get_info(self, cvs, name):
        url = "https://www.imdb.com/name/{}/?ref_=nmls_hd"
        with open(cvs, 'r') as f:
            file_content = f.readlines()
            file_content = file_content[1:]
            for line in file_content:
                if name in line:
                    id = line[1:12]
                    if len(id)== 11:
                        if ',' == id[10]:
                            id=id[1:10]
                        elif ','!= id[10]:
                            id=id[2:11]
                    url = "https://www.imdb.com/name/{}/?ref_=nmls_hd".format(id)
        r = requests.get(url)
        soup = BeautifulSoup(r.content,'html.parser')
        text1 = soup.find_all('div',{'class':'name-trivia-bio-text'})
        print(url)
        self.get_data(r.content)
        self.biography(cvs, name)
        self.all_time_movies(cvs, name)
        self.all_time_movies2(cvs, name)
        self.process_sub_links(cvs, name)
        self.average_movie_ratings(cvs, name)
        self.top_5(cvs, name)
        
        
    def get_data(self, content):
        soup = BeautifulSoup(content, 'html.parser')
        print(soup.title.string)
        list1 = []
        list2 = []
        item = ""
        for link in soup.find_all('a'):
            item = link.get('href')
            list1.append(item)

        self.get_sub_data(list1)
        return list1

    def get_sub_data(self, content):
        list2 = []
        list3 = []
        for item in content:
            if isinstance(item, str):
                list2.append(item)
                for value in list2:
                    if 'http' in value and 'url' not in value:
                        list3.append(value)
                        #print(value)
        return list3

    #biography
    def biography(self, cvs, name):
            with open(cvs, 'r') as f:
                    file_content = f.readlines()
                    file_content = file_content[1:]
                    for line in file_content:
                        if name in line:
                            id = line[1:12]
                            if len(id)== 11:
                                if ',' == id[10]:
                                    id=id[1:10]
                                elif ','!= id[10]:
                                    id=id[2:11]

                            url1 = 'https://www.imdb.com/name/{}/bio?ref_=nm_ov_bio_sm'.format(id)
            r = requests.get(url1)
            #print(r.content)
            soup = BeautifulSoup(r.content, 'html.parser')
            bio_title = soup.find('h1',{'class':'header'}).text
            print('\033[1m'+bio_title+'\033[0m')
            bio = soup.find('div',{'class':'soda odd'}).text
            print(bio)
            
    #awards
    def process_sub_links(self, cvs, name):
        with open(cvs, 'r') as f:
                file_content = f.readlines()
                file_content = file_content[1:]
                for line in file_content:
                    if name in line:
                        id = line[1:12]
                        if len(id)== 11:
                            if ',' == id[10]:
                                id=id[1:10]
                            elif ','!= id[10]:
                                id=id[2:11]

                        url1 = 'https://www.imdb.com/name/{}/awards?ref_=nm_awd'.format(id)
        r = requests.get(url1)
        #print(r.content)
        soup = BeautifulSoup(r.content, 'html.parser')
        awards_title = soup.find('h1').text
        print('\033[1m'+awards_title+'\033[0m')
        awards = soup.find_all('table',{'class':'awards'})
        for x in awards:
            print(x.get_text())
        soup.prettify()
        
    #all time movies    
    def all_time_movies(self, cvs, name):
        with open(cvs, 'r') as f:
                file_content = f.readlines()
                file_content = file_content[1:]
                for line in file_content:
                    if name in line:
                        id = line[1:12]
                        if len(id)== 11:
                            if ',' == id[10]:
                                id=id[1:10]
                            elif ','!= id[10]:
                                id=id[2:11]

                        url1 = 'https://www.imdb.com/filmosearch/?explore=title_type&role={}&ref_=filmo_ref_typ&sort=year,asc&mode=detail&page=1&title_type=movie'.format(id)
        r = requests.get(url1)
        #print(r.content)
        soup = BeautifulSoup(r.content, 'html.parser')
        title = soup.find('h1').text
        print('\033[1m'+title+'\033[0m')
        filmography = soup.find_all('div', class_="lister-item-content")
        for film in filmography:
            print(film.find('h3').text)
            print(film.find('p',{'class':'text-muted text-small'}).text)
            #print(film.find('strong').text)
            print(film.find('p',{'class':''}).text)
            
    def all_time_movies2(self, cvs, name):
        with open(cvs, 'r') as f:
                file_content = f.readlines()
                file_content = file_content[1:]
                for line in file_content:
                    if name in line:
                        id = line[1:12]
                        if len(id)== 11:
                            if ',' == id[10]:
                                id=id[1:10]
                            elif ','!= id[10]:
                                id=id[2:11]

                        url1 = 'https://www.imdb.com/filmosearch/?explore=title_type&role={}&ref_=filmo_ref_typ&sort=year,asc&mode=detail&page=2&title_type=movie'.format(id)
        r = requests.get(url1)
        #print(r.content)
        soup = BeautifulSoup(r.content, 'html.parser')
        filmography = soup.find_all('div', class_="lister-item-content")
        for film in filmography:
            print(film.find('h3').text)
            print(film.find('p',{'class':'text-muted text-small'}).text)
            #print(film.find('strong').text)
            print(film.find('p',{'class':''}).text)
    
    #top 5 movies
    def top_5(self, cvs, name):
        with open(cvs, 'r') as f:
                file_content = f.readlines()
                file_content = file_content[1:]
                for line in file_content:
                    if name in line:
                        id = line[1:12]
                        if len(id)== 11:
                            if ',' == id[10]:
                                id=id[1:10]
                            elif ','!= id[10]:
                                id=id[2:11]

                        url1 ='https://www.imdb.com/filmosearch/?explore=title_type&role={}&ref_=filmo_ref_typ&sort=moviemeter,asc&mode=detail&page=1&title_type=movie'.format(id)
        r = requests.get(url1)
        soup = BeautifulSoup(r.content, 'html.parser')
        title = soup.find('h1').text
        print('\033[1m'+title+'\033[0m')
        filmography = soup.find_all('div', class_="lister-item-content",limit=5)
        for film in filmography:
            print(film.find('h3').text)
            print(film.find('p',{'class':'text-muted text-small'}).text)
            print(film.find('p',{'class':''}).text)
    
    #average overall rating
    def average_movie_ratings(self, cvs, name):
        with open(cvs, 'r') as f:
                file_content = f.readlines()
                file_content = file_content[1:]
                for line in file_content:
                    if name in line:
                        id = line[1:12]
                        if len(id)== 11:
                            if ',' == id[10]:
                                id=id[1:10]
                            elif ','!= id[10]:
                                id=id[2:11]

                            url1 = 'https://www.imdb.com/filmosearch/?explore=title_type&role={}&ref_=filmo_ref_typ&sort=year,asc&mode=detail&page=1&title_type=movie'.format(id)
        r = requests.get(url1)
        #print(r.content)
        mylist = []
        soup = BeautifulSoup(r.content, 'html.parser')
        print('\033[1m'+"Averge Overall Movie Rating "+'\033[0m')
        filmography = soup.find_all('div', class_="lister-item-content")
        for film in filmography:
            year =film.find('span',class_="lister-item-year text-muted unbold").text
            rating =float(film.find('div',class_="inline-block ratings-imdb-rating").text)
            mylist.append(rating)
            
        average = str(sum(mylist)/len(mylist))
        print("Overall Average movie rating is "+ average)
        print("   ")
       
        #soup.prettify()
        
        
        

if __name__ == '__main__':
    file_csv = 'Top 50 Popular Hollywood Actors and Actresses.csv'
    #print(file_csv)
    da_ss = ActorsnActresses()
    data = da_ss.printingcsv(file_csv)
    #Give the name of the Actor or Actress
    celeb_name = input("Enter name of celeb from the list above:")
    full_name = celeb_name.title()
    print(full_name)
    data2 = da_ss.get_info(file_csv, full_name)
    #print (data)
```
