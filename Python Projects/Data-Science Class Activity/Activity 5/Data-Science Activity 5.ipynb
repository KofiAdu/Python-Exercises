## Exercise Question 2: Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this.
![Image 1](images/e2.png)

```python
# Excercise 2

#Start Code 


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

df= pd.read_csv("company_sales_data.csv")
df

plt.plot(df["month_number"],df["total_profit"])
x=df['month_number']
y=df['total_profit']
plt.plot(x,y,'b-','go',linewidth=3,color='red',marker='o')

plt.xlim((1,12))
plt.ylim((100000,500000))


plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Company Sales Data")


csd = mpatches.Patch(color='red', label='Profit Data of Last Year',)
plt.legend(handles=[csd],loc='lower right')



#end code
```

## Exercise Question 3: Read all product sales data and show it  using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

The graph should look like this.

![Image3](images/e3.png)

```python
# Excercise 3

#Start Code 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df= pd.read_csv("company_sales_data.csv")
df

tp=df[['toothpaste']]
fc=df[['facecream']]
fw=df[['facewash']]
sh=df[['shampoo']]
bs=df[['bathingsoap']]
m =df[['moisturizer']]

x=df[['month_number']]


plt.plot(x,tp,'g',label="Toothpaste Sales Data",lw=3,marker='o')
plt.plot(x,fc,'b',label="Facecream Sales Data",lw=3,marker='o')
plt.plot(x,fw,'brown',label="Facewash Sales Data",lw=3,marker='o')
plt.plot(x,sh,'purple',label="Shampoo Sales Data",lw=3,marker='o')
plt.plot(x,bs,'r',label="Bathingsoap Sales Data",lw=3,marker='o')
plt.plot(x,m,'v',label="Moisturizer Sales Data",lw=3,marker='o')

plt.xlabel("Months")
plt.ylabel("Sales units in numbers")
plt.title("Sales")

plt.legend(('Toothpaste Sales Data','Facecream Sales Data',"Facewash Sales Data","Shampoo Sales Data","Bathingsoap Sales Data"))

#end code
```

## Exercise Question 9: Read Bathing soap facewash of all months and display it using the Subplot

The Subplot should look like this.

![image 9](images/e9.png)

```python
# Excercise 9

#Start Code 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df= pd.read_csv("company_sales_data.csv")
df
bs=df[['bathingsoap']]
fw=df[['facewash']]

fig=plt.figure()

axis1= fig.add_subplot(211)
plt.plot(bs,linewidth=3,marker='o')
axis1.set_title("Sales Data for Bathing Soap")


axis2= fig.add_subplot(212)
plt.plot(fw,lw=3,marker='o')
axis2.set_title("Sales Data for Facewash")

#end code
```
