```python
import pandas as pd
import numpy as np

file_name= "bill_of_rights.txt"

num_lines=0
with open(file_name, "r")as file:
    for lines in file:
        num_lines+=1
        
print(num_lines)

    
```

```python
file_name= "bill_of_rights.txt"

num_words=0
with open(file_name, "r")as file:
    for lines in file:
        words=lines.split()
        
        num_words += len(words)
        
print(num_words)
```

```python
import collections

import re

words=re.findall(r'\w+',open("bill_of_rights.txt").read().lower())
most_freq=collections.Counter(words).most_common(1)
print(most_freq)

```

```python
x=open("stats.txt","w")
x.write(f'Number of lines is {num_lines} \n')
x.write(f'Number of words is {num_words} \n')
x.write(f'Most frequently used word is {most_freq} \n')
x.close()
```
