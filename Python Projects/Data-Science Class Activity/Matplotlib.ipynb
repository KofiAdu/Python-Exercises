```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,100,1000)
y = np.power(x,2)
plt.plot(x,y)
plt.savefig("lineplot.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,100,1000)
y = np.power(x,2)
plt.plot(x,y)


#addlimits plt

plt.xlim((1,50))
plt.ylim((0,5000))

#labels
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.title("title")
plt.savefig("lineplot1.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,100,1000)
y = np.power(x,2)
plt.plot(x,y)

z = np.power(x,3)
plt.plot(x,y,'b-',x,z,'go',linewidth=3)
#addlimits plt

plt.xlim((1,50))
plt.ylim((0,5000))

#labels
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.title("title")

#add legend
plt.legend(('$x^2$','$x^3$'))

plt.savefig("lineplot2.png")
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1= np.arange(0.0,5.0,0.1)
t2= np.arange(0.0,5.0,0.02)

#create canvas for fig

fig=plt.figure()

#subplot

axis1= fig.add_subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')
axis1.set_title("plot 1")

axis2= fig.add_subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r-')
axis2.set_title("plot 2")
```
